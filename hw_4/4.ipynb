```python
import numpy as np
from scipy.misc import derivative
from matplotlib import pyplot as plt
```

# Уровень 0
Посчитайте значение первой и второй производной функции в точке x = 1.

```python
def f(x):
    return x**5 + 4*np.sin(2*x) + np.cos(3*x+3)
f(1)
```

```python
f_1 = derivative(f, 1.0)
f_1
```

```python
f_2 = derivative(f, 1.0, n=2)
f_2
```

# Уровень 1
Постройте график первой и второй производной функции на отрезке [-5; 5].

```python
def f2(y):
    return np.sin(2*y+1)**5
```

```python
y = np.linspace(-5, 5)
fy=f2(y)
fy_1 = [derivative(f2,yy,n=1) for yy in y] #первая производная 
fy_2 = [derivative(f2,yy,n=2) for yy in y] #вторая производная
```

```python
plt.plot(y,fy) # исходная функция
```

```python
plt.plot(y,fy_1, label = '1 derivative')
plt.plot(y,fy_2, label = '2 derivative')
plt.legend()
plt.show()
```
